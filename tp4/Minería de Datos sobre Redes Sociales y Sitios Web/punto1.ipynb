```python
import tweepy
import json
```

```python
consumer_key=""
consumer_secret=""
access_token=""
access_token_secret=""

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
```

```python
## a. Los tres trends más RT del momento.

trends1 = api.trends_place(23424747)
trends = trends1[0]['trends']
for i in range(3):   
    print(trends[i]['name'])
```

```python
## b. Listar nombres de usuarios que publicaron con los hashtag de los trends del punto anterior. 
data = []
for i in range(3):
    print("TREND", trends[i]['name'])
    for tweet in tweepy.Cursor(api.search, q=trends[i]['name']).items(200):
        user = tweet._json['user']
        print("usuario: ", user['screen_name'], "nombre: ", user['name'])
        data.append({trends[i]['name']: tweet._json})
        
with open('tweets.json', 'w') as file:
    json.dump(data, file, indent=4)
```

```python
## c. ¿Cuáles son las cinco palabras más utilizadas en los tweets del primer ítem?
import operator
palabras_omitir = ['ya','el', 'la','lo', 'los', 'las', 'un', 'una', 'unos', 'unas','a', 'ante',
                  'bajo','cabe','como','con','contra','de','desde','durante','en','entre','excepto',
                  'hacia','hace','hasta','mediante','para','por','segun','sin','sobre','tras',
                  'del','que','se','muchos','muchas','mucho','mucha','hace','y','o','e','Mucho',
                  'han','ha','he','Y','al','ello','conlleva','todos','todo','todas','toda','es',
                   'sea','Muchas','algunas','algunos','pero','no','si','pocas','“Una','da','-y','más','siguen',
                   'parte','esta','Pero','otra','otro','casi','ninguna','siendo','Este','también',
                  'sino','sido','sigue','siendo-','aún','mayor','habría','Las','vez','menos','gran',
                  'cuales','ser','Por','ni','En','últimas','basado','Se','como','su','además','sólo',
                  'sus','misma','antes','esta.','deben','misma','RT']


with open('tweets.json') as file:
    data = json.load (file)
    
    for i in range(3):
        texto = ""
        for tweet in data:
            for k in tweet:
                if k == trends[i]['name']:
                    texto += tweet[k]['text'] + '\n'
        palabras_sin_omitir = texto.split()
        palabras = []
        for w in palabras_sin_omitir:
            if w not in palabras_omitir:
                palabras.append(w)
        
        top_palabras = {}
        print("Trend", trends[i]['name'])
        for w in palabras:
            if w not in top_palabras:
                cant = palabras.count(w)
                top_palabras[w] = cant
        top_palabras_sort = sorted(top_palabras.items(), key=operator.itemgetter(1), reverse=True)
        for i in range(5):
            print(top_palabras_sort[i])
```

```python
## d. Listar los primeros 10 usuarios con mayor cantidad de seguidores. 

with open('tweets.json') as file:
    data = json.load (file)
    
    datos_usuarios = []
    for tweet in data:
        for k in tweet:
            datos_usuarios.append({'usuario': tweet[k]['user']['screen_name'],
                                   'nombre': tweet[k]['user']['name'],
                                   'followers':tweet[k]['user']['followers_count']})
    top_users = sorted(datos_usuarios ,key=lambda k: int(k['followers']),reverse=True)
    for i in range(10):
        print("usuario ",top_users[i]['usuario'],"nombre ",top_users[i]['nombre'],
             "followers ",top_users[i]['followers'])
```

```python
## e. Listar la ubicación (o en su defecto time-zone) del tweet.

with open('tweets.json') as file:
    data = json.load (file)
    
    datos_usuarios = []
    for tweet in data:
        for k in tweet:
            datos_usuarios.append({'usuario': tweet[k]['user']['screen_name'],
                                   'nombre': tweet[k]['user']['name'],
                                   'followers':tweet[k]['user']['followers_count'],
                                    'ubicacion':tweet[k]['user']['location'],
                                  'time-zone':tweet[k]['user']['time_zone']})
    top_users = sorted(datos_usuarios ,key=lambda k: int(k['followers']),reverse=True)
    for i in range(10):
        print("usuario ",top_users[i]['usuario'],"nombre ",top_users[i]['nombre'],
             "followers ",top_users[i]['followers'],"ubicacion ",top_users[i]['ubicacion'],
             "time-zone ",top_users[i]['time-zone'])
```

```python
## f. Listar los cinco tweets más populares. 

with open('tweets.json') as file:
    data = json.load (file)
    
    datos_usuarios = []
    for tweet in data:
        for k in tweet:
            datos_usuarios.append({'usuario': tweet[k]['user']['screen_name'],
                                   'nombre': tweet[k]['user']['name'],
                                   'text':tweet[k]['text'],
                                  'fav':tweet[k]['user']['favourites_count']})
    top_users = sorted(datos_usuarios ,key=lambda k: int(k['fav']),reverse=True)
    for i in range(5):
        print("usuario ",top_users[i]['usuario'],"nombre ",top_users[i]['nombre'],
             "texto",top_users[i]['text'],"favourites_count",top_users[i]['fav'])
```

```python
## g. Listar a los seguidores del autor del tweet más popular. 

with open('tweets.json') as file:
    data = json.load (file)
    
    datos_usuarios = []
    for tweet in data:
        for k in tweet:
            datos_usuarios.append({'usuario': tweet[k]['user']['screen_name'],
                                   'nombre': tweet[k]['user']['name'],
                                   'text':tweet[k]['text'],
                                  'fav':tweet[k]['user']['favourites_count'],
                                  'user_mentions':tweet[k]['entities']['user_mentions']})
    top_users = sorted(datos_usuarios ,key=lambda k: int(k['fav']),reverse=True)
    for i in range(1):
        print("usuario ",top_users[i]['usuario'],"nombre ",top_users[i]['nombre'],
             "texto",top_users[i]['text'],"favourites_count",top_users[i]['fav'])
        print("seguidores ")
        user_mentions = top_users[i]['user_mentions']
        for u in user_mentions:
            print("usuario ", u['screen_name'])
```

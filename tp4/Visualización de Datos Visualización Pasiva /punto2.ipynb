```python
import folium
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def elegir_color(casos):
    if casos < 5000:
        return '#31cc65'
    if casos < 10000:
        return '#3186cc'
    if casos < 15000:
        return '#ffff63'
    if casos < 100000: 
        return '#ff1c1c'
```

```python
## a. Indicar la cantidad de casos con un círculo de diferente tamaño.
## b. Indicar la penetración de casos por provincia utilizando una gama de colores para cada círculo.

archivo = pd.read_csv('covid_arg_transformado.csv')
archivo2 = pd.read_csv('pop_provs.csv')
casos_prov=pd.merge(archivo,archivo2,left_on='provincia',right_on='name')

casos_prov.drop(['Unnamed: 0_x','dia_inicio','ciudad','nue_casosconf_diff','tot_fallecidos','nue_fallecidos_diff',
                'transmision_tipo','informe_tipo','informe_link','Unnamed: 0_y','code','name','pop_dens','country_percentage'],axis='columns',inplace=True)
casos_prov['casos_cada_100mil_hab'] = casos_prov['tot_casosconf'] / (casos_prov['total_pop'] / 100000)

provincias = casos_prov.groupby('provincia')
provincias_casos = []
for p in provincias:
    provincias_casos.append({'provincia':p[0],
                             'total_casos':p[1].loc[p[1].index[-1], 'tot_casosconf'],
                            'total_casos_cada_100mil_hab':p[1].loc[p[1].index[-1], 'casos_cada_100mil_hab']})


m = folium.Map(location=[-38.416097, -63.616672],zoom_start=4)

for p in provincias_casos:

    if p['provincia'] == 'Buenos Aires':
        folium.CircleMarker(
            location=[ -34.6083, -58.3712],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color= elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color= elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    
    if p['provincia'] == 'Chubut':
        folium.CircleMarker(
            location=[ -44, -69],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Rio Negro':
        folium.CircleMarker(
            location=[ -40.8, -63],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Santa Cruz':
        folium.CircleMarker(
            location=[-50.1333300,-68.3333300],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Tierra del Fuego':
        folium.CircleMarker(
            location=[-54.5000000, -67.0000000],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Neuquen':
        folium.CircleMarker(
            location=[-38.9516100,-68.0591000],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Córdoba':
        folium.CircleMarker(
            location=[  -31.417, -64.183],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'La Pampa':
        folium.CircleMarker(
            location=[-36.61667,-64.28333 ],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Santa Fe':
        folium.CircleMarker(
            location=[-31.6333300,-60.7000000],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Mendoza':
        folium.CircleMarker(
            location=[-32.8903 , -68.8472 ],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'La Rioja':
        folium.CircleMarker(
            location=[ -29.4131, -66.8558 ],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'San Luis':
        folium.CircleMarker(
            location=[-33.2950100,-66.3356300],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'San Juan':
        folium.CircleMarker(
            location=[ -31.5372,-68.5253 ],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Corrientes':
        folium.CircleMarker(
            location=[-27.4806000,-58.8341000],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Misiones':
        folium.CircleMarker(
            location=[-26.61667, -54.13333],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Tucuman':
        folium.CircleMarker(
            location=[-26.8241400,-65.2226000],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Santiago del Estero':
        folium.CircleMarker(
            location=[ -27.79511,-64.26149 ],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Chaco':
        folium.CircleMarker(
            location=[-25.94679,-60.61947],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Formosa':
        folium.CircleMarker(
            location=[-26.1775300,-58.1781400],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Salta':
        folium.CircleMarker(
            location=[ -24.7859,-65.41166 ],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Jujuy':
        folium.CircleMarker(
            location=[ -24.19457,-65.29712],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Entre Ríos':
        folium.CircleMarker(
            location=[ -30.74179,-59.64517],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
    if p['provincia'] == 'Catamarca':
        folium.CircleMarker(
            location=[-28.01009,-65.70874],
            radius=p['total_casos']/5000,
            popup=p['provincia'],
            color=elegir_color(p['total_casos_cada_100mil_hab']),
            fill=True,
            fill_color=elegir_color(p['total_casos_cada_100mil_hab'])
        ).add_to(m)
m
```

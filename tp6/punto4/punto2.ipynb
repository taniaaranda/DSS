```python
from skimage.io import imread 
from skimage.color import rgb2gray 
import numpy as np 
import matplotlib.pyplot as plt 
from scipy import ndimage
from sklearn.cluster import KMeans
```

```python
# Escalar los valores de pxeles de la imagen dentro de 0-1 
img = imread('iris.jpg') / 255
plt.imshow(img) 
plt.title('Original') 
plt.show()
```

```python
# Para agrupar la imagen usando k-means, primero hay que convertirla en una matriz bidimensional 
image_2D = img.reshape (img.shape [0] * img.shape [1], img.shape [2])

# ajustar el tamao del clster  
kmeans = KMeans (n_clusters = 80, random_state = 0).fit(image_2D) 
clustered = kmeans.cluster_centers_[kmeans.labels_]

# Remodelar la imagen de 2D a 3D 
clustered_3D = clustered.reshape(img.shape [0], img.shape [1], img.shape [2])
plt.imshow(clustered_3D) 
plt.title('Imagen agrupada') 
plt.show()
```

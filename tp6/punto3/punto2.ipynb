```python
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn import svm, datasets 
import pandas as pd
from sklearn.svm import SVC
from sklearn import metrics
from sklearn.metrics import f1_score, accuracy_score, precision_score
```

**Kernel lineal**

```python

iris = datasets.load_iris() 
# Tomar las primeras dos caracteristicas 
  
X = iris.data[:, :2] 
y = iris.target

# C es el parametro de regularizacion de SVM 
C = 1.0 
  
# Crear una instancia de SVM y ajustar los datos.  
svc = svm.SVC(kernel ='linear', C = 1).fit(X, y) 
  
# Crear linea 
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
h = (x_max / x_min)/100
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), 
         np.arange(y_min, y_max, h)) 
  
# Trazar los datos 
plt.subplot(1, 1, 1) 
  
# Predecir el resultado dando datos al modelo 
Z = svc.predict(np.c_[xx.ravel(), yy.ravel()]) 
Z = Z.reshape(xx.shape) 
plt.contourf(xx, yy, Z, cmap = plt.cm.Paired, alpha = 0.8) 
  
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Paired) 
plt.xlabel('Longitud de Sepa') 
plt.ylabel('Ancho de Sepa') 
plt.xlim(xx.min(), xx.max()) 
plt.title('kernel lineal iris') 
  

plt.show() 
```

**Kernel rbf**

```python
#instanciar la clase SVC con kernel rbf 
#con gamma 0.01, el límite de decisión no es muy 'curvo', sino que es solo un gran arco de barrido.
svm = SVC(kernel='rbf', random_state=0, gamma=.01, C=1)
svm.fit(X, y)
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]) 
Z = Z.reshape(xx.shape) 
plt.contourf(xx, yy, Z, cmap = plt.cm.Paired, alpha = 0.8) 
  
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Paired) 
plt.xlabel('Longitud de Sepa') 
plt.ylabel('Ancho de Sepa') 
plt.xlim(xx.min(), xx.max()) 
plt.title('kernel rbf iris') 
  

plt.show() 
```

```python
#con gammaa 1 el límite de decisión comienza a cubrir mejor la distribución de los datos
svm = SVC(kernel='rbf', random_state=0, gamma=1, C=1)
svm.fit(X, y)

Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]) 
Z = Z.reshape(xx.shape) 
plt.contourf(xx, yy, Z, cmap = plt.cm.Paired, alpha = 0.8) 
  
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Paired) 
plt.xlabel('Longitud de Sepa') 
plt.ylabel('Ancho de Sepa') 
plt.xlim(xx.min(), xx.max()) 
plt.title('kernel rbf iris') 
  

plt.show() 
```

```python
#Con gamma 10 la propagación del núcleo es menos pronunciada
#El límite de decisión comienza a verse afectado en gran medida por puntos de datos individuales (varianza).
svm = SVC(kernel='rbf', random_state=0, gamma=10, C=1)
svm.fit(X, y)
Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]) 
Z = Z.reshape(xx.shape) 
plt.contourf(xx, yy, Z, cmap = plt.cm.Paired, alpha = 0.8) 
  
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Paired) 
plt.xlabel('Longitud de Sepa') 
plt.ylabel('Ancho de Sepa') 
plt.xlim(xx.min(), xx.max()) 
plt.title('kernel rbf iris') 
  

plt.show() 
```

```python
#Con gamma alto el límite de decisión depende casi por completo de puntos de datos individuales
#creando "islas" Estos datos están claramente sobreajustados.
svm = SVC(kernel='rbf', random_state=0, gamma=100, C=1)
svm.fit(X, y)

Z = svm.predict(np.c_[xx.ravel(), yy.ravel()]) 
Z = Z.reshape(xx.shape) 
plt.contourf(xx, yy, Z, cmap = plt.cm.Paired, alpha = 0.8) 
  
plt.scatter(X[:, 0], X[:, 1], c = y, cmap = plt.cm.Paired) 
plt.xlabel('Longitud de Sepa') 
plt.ylabel('Ancho de Sepa') 
plt.xlim(xx.min(), xx.max()) 
plt.title('kernel rbf iris') 
  

plt.show() 
```

**Cuantificación de la calidad de las predicciones**

```python
svc_y_pred = svc.predict(X)
svm_y_pred = svm.predict(X)
```

**f1 score**

```python
print(" métricas para cada etiqueta y encuentre su media no ponderada")
print("f1_score kernel lineal:",f1_score(y, svc_y_pred, average='macro'))
print("f1_score kernel rbf:",f1_score(y, svm_y_pred, average='macro'))
```

```python
print(" métricas globalmente contando el total de verdaderos positivos, falsos negativos y falsos positivos.")
print("f1_score kernel lineal:",f1_score(y, svc_y_pred, average='micro'))
print("f1_score kernel rbf:",f1_score(y, svm_y_pred, average='micro'))
```

```python
print("métricas para cada etiqueta y encuentra su promedio ponderado por soporte (el número de instancias reales para cada etiqueta)")
print("f1_score kernel lineal:",f1_score(y, svc_y_pred, average='weighted'))
print("f1_score kernel rbf:",f1_score(y, svm_y_pred, average='weighted'))
```

```python
print("se devuelven las puntuaciones de cada clase")
print("f1_score kernel lineal:",f1_score(y, svc_y_pred, average=None))
print("f1_score kernel rbf:",f1_score(y, svm_y_pred, average=None))
```

**accuracy**

```python
print("precisión del subconjunto: el conjunto de etiquetas predichas para una muestra debe coincidir exactamente con el conjunto de etiquetas correspondiente")
print("accuracy_score kernel lineal: ",accuracy_score(y, svc_y_pred))
print("accuracy_score kernel rbf: ",accuracy_score(y, svm_y_pred))
```

**precision**

```python
print("métricas para cada etiqueta y encuentre su media no ponderada")
print("presicion_score kernel lineal: ",precision_score(y, svc_y_pred,average='macro'))
print("precision_score kernel rbf: ",precision_score(y, svm_y_pred,average='macro'))
```

```python
print("métricas globalmente contando el total de verdaderos positivos, falsos negativos y falsos positivos")
print("presicion_score kernel lineal: ",precision_score(y, svc_y_pred,average='micro'))
print("precision_score kernel rbf: ",precision_score(y, svm_y_pred,average='micro'))
```

```python
print("métricas para cada etiqueta y encuentra su promedio ponderado por soporte")
print("presicion_score kernel lineal: ",precision_score(y, svc_y_pred,average='weighted'))
print("precision_score kernel rbf: ",precision_score(y, svm_y_pred,average='weighted'))
```

```python
print("se devuelven las puntuaciones de cada clase")
print("presicion_score kernel lineal: ",precision_score(y, svc_y_pred,average=None))
print("precision_score kernel rbf: ",precision_score(y, svm_y_pred,average=None))
```

**compute area under the curve**

```python
print("el área bajo la curva (AUC) usando la regla trapezoidal")
print("presicion_score kernel lineal: ",metrics.auc(y, svc_y_pred))
print("precision_score kernel rbf: ",metrics.auc(y, svm_y_pred))
```

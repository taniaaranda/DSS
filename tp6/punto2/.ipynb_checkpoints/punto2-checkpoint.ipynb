**Regresion**

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from six import StringIO  
from IPython.display import Image  
from sklearn.tree import export_graphviz
import pydotplus
from pydotplus import graph_from_dot_data
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#carga de datos
dataset = pd.read_excel('Concrete_Data.xls')
dataset
```

```python
X = dataset.iloc[:, 0:7].values
y = dataset.iloc[:, 8].values

# Ajuste del arbol de decision al dataset
regressor = DecisionTreeRegressor(max_depth = 4)
regressor.fit(X, y)

x_labels = ['cement','Blast Furnace Slag','Fly Ash','water','Superplasticizer','Coarse Aggregate',
            'Fine Aggregate']
dot_data = StringIO()
export_graphviz(regressor, out_file = dot_data,
                feature_names = x_labels,
                class_names = str(y),
                filled = True, rounded = True,
                special_characters = True)
graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
Image(graph.create_png())
```

**CuantificaciÃ³n de la calidad de las predicciones**


```python
y_pred= regressor.predict(X)
y_pred
```

**mean squared error**

```python
print (u'Error cuadrÃ¡tico medio: %.2f' \
        % mean_squared_error(y, y_pred))
```

**r2**

```python
print (u'EstadÃ­stico R_2: %.2f' % r2_score(y, y_pred))
```
